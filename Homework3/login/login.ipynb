```python
import base64

def decodeBase64(input):
    try:
    
        while len(input) % 4 != 0:
            input += '='
                
        decoded_bytes = base64.b64decode(input)     
        decoded_text = decoded_bytes.decode('utf-8')
        return decoded_text
    
    except Exception as e:
        return print ('Error while decoding the text') 


encoded_string_1 = "cGljb0NURns1M3J2M3JfNTNydjNyXzUzcnYzcl81M3J2M3JfNTNydjNyfQ"
encoded_string_2 = "YWRtaW4"

decoded_string1 = decodeBase64(encoded_string_1)
decoded_string2 = decodeBase64(encoded_string_2)

print('This is the flag:',decoded_string1)
print("The password for the admin account is", decoded_string2)
```
